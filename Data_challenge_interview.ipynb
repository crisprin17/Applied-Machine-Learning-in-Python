###Sympson paradox
Simpson's Paradox is a classic pitfall in data analysis that can mislead conclusions if not properly addressed.
**Always explore subgroup behavior. Always stratify by potential confounders (e.g., time, groups)**
Definition:

When a trend appears in different groups of data but reverses when these groups are combined.

Can affect time series, especially when:
* Trends vary over time.
* Aggregating across periods hides subgroup behavior.
* A seasonal effect masks or reverses overall trends.
* watch out for hidden temporal confounders (e.g., policy changes, external events).

Example:
Drug A appears better than Drug B overall, but within age groups, Drug B is better.


How to Spot Simpson's Paradox


1.  Aggregate your data and look for trends.
2.  Stratify by a potential confounding variable (e.g., gender, time period, location).
3. Compare within-group trends vs. overall trends.
4. Stratify by time, category, or any suspected confounder.
5. Visualize both overall and group-specific trends!

How to Correct for Simpson's Paradox
Techniques:
* Stratified Analysis: Report results per group.
* Multivariate Models: Include confounders in your regression.
* Causal Inference: Use methods like propensity score matching, instrumental variables.
* Use multivariate models to control for group effects.

Key Takeaways:
Simpson's Paradox = Watch out for misleading aggregated trends.

### 1. Handling Missing Values
Why Important?
* Missing values can bias results or cause errors in modeling.
* Some models (e.g., XGBoost) handle missing data; others (e.g., linear models) don't.

What to Notice?

* Is it MCAR (Missing Completely At Random), MAR, or MNAR?

* Is the missingness informative?

Key Actions:
* Drop rows/columns with too much missingness.
* Impute values (mean, median, mode, KNN, interpolation, MICE).
REMEMBER: Imputation should preserve distribution and relationship.

MICE imputation, or Multiple Imputation by Chained Equations, is a method for handling missing data in datasets by using multiple iterations of machine learning models to predict and fill in missing values.
How MICE Works:
1. Identify Missing Data:
The process begins by identifying all the missing values in the dataset.
2. Iterative Imputation:
MICE uses an iterative approach where each variable with missing data is imputed using predictive models.
3. Predictive Models:
In each iteration, a predictive model (e.g., a regression model) is built to estimate missing values in one variable using the non-missing values of other variables in the dataset.
4. Chained Equations:
The "chained equations" part refers to the sequential nature of the imputation process. Each variable is imputed based on the results of previous imputations, creating a chain of predictions.
5. Multiple Imputations:
MICE generates multiple imputed datasets (e.g., 5 or 10) by repeating the imputation process several times with different random starting points. This introduces uncertainty into the imputed values, making the results more robust.
6. Statistical Analysis:
Once multiple imputed datasets are created, standard statistical analyses are performed on each dataset.
7. Pooling Results:
The results from the different imputed datasets are then combined (pooled) to provide a final estimate of the statistical analysi

```python
import plotly.express as px
import pandas as pd

# Visualize missing data pattern
missing_df = df.isnull().sum().reset_index()
missing_df.columns = ['column', 'missing_count']
missing_df = missing_df[missing_df['missing_count'] > 0]

fig = px.bar(missing_df, x='column', y='missing_count', title='Missing Values per Column')
fig.show()

# Drop columns with > 50% missing
threshold = 0.5
df = df.loc[:, df.isnull().mean() < threshold]

# Impute numeric with median
for col in df.select_dtypes(include='number').columns:
    df[col].fillna(df[col].median(), inplace=True)

# Impute categorical with mode
for col in df.select_dtypes(include='object').columns:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

###2. Removing Duplicates

Why Important?
* Duplicates distort statistics, create bias, and may overfit models. Duplicates inflate the importance of specific records Always check for the exact and contextual duplicates

What to Notice?
* Are they exact duplicates or partial (e.g., same user, different timestamps)?

Key Actions:
* Remove exact or partial duplicates.

```python
# Exact duplicates
duplicate_count = df.duplicated().sum()
print(f"Exact duplicate rows: {duplicate_count}")
df = df.drop_duplicates()

# Partial duplicate check (e.g., user_id + date)
if 'user_id' in df.columns and 'timestamp' in df.columns:
    df['date'] = pd.to_datetime(df['timestamp']).dt.date
    partial_duplicates = df.duplicated(subset=['user_id', 'date'])
    print(f"Partial duplicates (user_id + date): {partial_duplicates.sum()}")
    df = df[~partial_duplicates]
```

What to Notice?
* Visual outliers.
* Extreme values in boxplots, histograms.

Key Actions:
* Decide whether to remove, cap, or treat. Capping is safer if some outlier are expected but extreme

```python
import plotly.graph_objects as go

# Example: Visualize outliers in numerical features
for col in df.select_dtypes(include='number').columns:
    fig = go.Figure()
    fig.add_trace(go.Box(y=df[col], name=col, boxpoints='all', jitter=0.3))
    fig.update_layout(title=f'Boxplot of {col}')
    fig.show()

# Capping outliers (e.g., using IQR method)
for col in df.select_dtypes(include='number').columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df[col] = df[col].clip(lower, upper)
```

###6. Normalizing Data: Log Transformations
Why Important?
* Helps normalize skewed distributions.
* Stabilizes variance, especially for income, prices, counts.
* Log transforms help in models assuming normality or linear relationships.

Stabilizes variance means transforming the data so that the spread or variability across different values is consistent, regardless of the value of the variable. In simpler terms, it involves making the "noise" or random variation in the data consistent across the board, rather than having it depend on the magnitude of the variable. This is particularly important when the variance isn't constant, as it can lead to problems in statistical analysis and modeling. (heteroschedacity)
* Many statistical methods, like regression analysis and analysis of variance (ANOVA), assume that the variance is constant across the data.
* Constant variance can improve the stability and accuracy of models. When the variance is consistent, the model can learn more reliably from the data and make more accurate predictions.
* Heteroscedasticity can lead to biased estimates and unreliable standard errors

What to Notice?
* Is the feature positively skewed?
* Any long tails?

Key Actions:
* Apply np.log1p() to compress scale.


Box-Cox transformation is used to make data more normally distributed, which can improve the performance of certain statistical models and algorithms that assume normality.
* Normality:
Many statistical techniques, like regression analysis, rely on the assumption that the data is normally distributed. If your data is skewed or has other non-normal distributions, applying Box-Cox can help it approximate a normal distribution.
* Variance Stabilization:
Box-Cox transformation can also help stabilize the variance of your data, making it more suitable for certain models that assume constant variance.


Models That Assume Normality:
* Linear Regression: Assumes normality of residuals, not necessarily features.

* Logistic Regression: Does not require normality.

* LDA (Linear Discriminant Analysis): Assumes normality of features within each class.

* Naive Bayes: Assumes features follow Gaussian distribution (Gaussian NB variant).

* T-tests / ANOVA: Assume normality for valid inference.

##2. Cross-Validation (CV)
The basic idea of cross validation is to split the data into training and validation sets, and then train the model on the training set and evaluate its performance on the validation set. This process is repeated multiple times, with different subsets of the data used for training and validation, and the average performance is calculated.
What is Cross-Validation?
* K-Fold CV splits data into k parts, trains on k-1, tests on the rest, repeats.
* Helps ensure stable performance, detects variance.

Why Important?
* Reduces risk of overfitting to one random train/test split.
* Gives better performance estimates.
* Can be used to compare differnt model or algorithm

Key Types:
* K-Fold CV: Standard for most tasks.
* Stratified K-Fold: For imbalanced classes.
* TimeSeriesSplit: For temporal data.

Best Practices:
* Use 5 or 10 folds.
* Combine CV with GridSearch or RandomSearch for hyperparameter tuning.


Hyperparameters are important because they can have a significant impact on the performance of a machine learning model. If the hyperparameters are set poorly, the model may not perform well, even if it is trained on a large amount of data.

* Grid search is a simple method for hyperparameter tuning, in which a grid of hyperparameter values is defined and the model is trained and evaluated for each combination of hyperparameters in the grid. The combination of hyperparameters that results in the best performance is then selected as the optimal set of hyperparameters.

* Random search is another method for hyperparameter tuning, in which hyperparameters are randomly sampled from a specified distribution and the model is trained and evaluated for each set of sampled hyperparameters. The set of hyperparameters that results in the best performance is then selected as the optimal set of hyperparameters.

* Bayesian optimization is a more advanced method for hyperparameter tuning, in which a probabilistic model is used to model the relationship between the hyperparameters and the model's performance. The model is updated as new evaluations are made, and the next set of hyperparameters to be evaluated is selected based on the current model.



**For hyperparameter tuning, you can use nested cross-validation. **
* Inner Loop: Use one cross-validation loop for hyperparameter tuning, evaluating performance with different hyperparameter settings.

* Outer Loop: Use another cross-validation loop to evaluate the performance of the model with the best hyperparameter settings found in the inner loop.


```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=42)
cv_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='accuracy')

print(f"Cross-validation accuracy scores: {cv_scores}")
print(f"Mean CV accuracy: {cv_scores.mean():.4f}, Std: {cv_scores.std():.4f}")



from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [5, 10, None],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
}

rf = RandomForestClassifier(random_state=42)
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, scoring='f1', n_jobs=-1, verbose=1)

grid_search.fit(X_train, y_train)
print(f"Best params: {grid_search.best_params_}")
print(f"Best F1 score: {grid_search.best_score_:.4f}")

```

```python
from sklearn.model_selection import GridSearchCV, cross_val_score, KFold
from sklearn.ensemble import RandomForestClassifier

# Outer loop CV unbiased model evaluation
outer_cv = KFold(n_splits=5, shuffle=True, random_state=42)

# Inner loop: Grid Search for parameter tuning
param_grid = {'n_estimators': [100, 200], 'max_depth': [5, 10, None]}
rf = RandomForestClassifier(random_state=42)
inner_cv = KFold(n_splits=3, shuffle=True, random_state=42)

grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=inner_cv, scoring='f1')

# Nested CV
nested_scores = cross_val_score(grid_search, X, y, cv=outer_cv, scoring='f1')

print(f'Nested CV F1 score: {nested_scores.mean():.4f} +/- {nested_scores.std():.4f}')
```

###3. Overfitting vs Underfitting
Definitions:
* Overfitting: Model learns noise, performs well on train, bad on test.
* Underfitting: Model is too simple, poor on both train and test.

What to Notice?
* Large gap between train and test accuracy = overfitting.
* Low accuracy on both = underfitting.

Best Practices:
* Use regularization (L1/L2).
* Limit model complexity.
* Use more data or feature engineering.

```python
# Example: Visualizing overfitting
train_scores = []
test_scores = []
depths = range(1, 20)

for d in depths:
    model = RandomForestClassifier(max_depth=d, random_state=42)
    model.fit(X_train, y_train)
    train_scores.append(model.score(X_train, y_train))
    test_scores.append(model.score(X_test, y_test))

import plotly.graph_objects as go
fig = go.Figure()
fig.add_trace(go.Scatter(x=list(depths), y=train_scores, mode='lines+markers', name='Train Accuracy'))
fig.add_trace(go.Scatter(x=list(depths), y=test_scores, mode='lines+markers', name='Test Accuracy'))
fig.update_layout(title='Overfitting vs Underfitting', xaxis_title='Model Depth', yaxis_title='Accuracy')
fig.show()
```

## SHAP Values

 a method for explaining machine learning model predictions by assigning each feature a contribution value. These values indicate how much each feature impacted the prediction for a particular instance. A positive SHAP value means a feature increased the prediction, while a negative value means it decreased the prediction. The magnitude of the SHAP value represents the strength of the feature's impact.


 SHAP deconstructs a prediction into a sum of contributions from each of the model's input variables.[1, 2] For each instance in the data (i.e. row), the contribution from each input variable (aka "feature") towards the model's prediction will vary depending on the values of the variables for that particular instance.

 Screen Shot 2025-04-27 at 11.36.42 AM.png

 For regression models, the base value is equal to the mean of the target variable (e.g. the mean house price in the dataset), whereas for classification models, the base value is equal to the prevalence of the positive class (e.g. the percentage of tumours in the dataset that are malignant)

Screen Shot 2025-04-27 at 11.43.13 AM.png


https://www.aidancooper.co.uk/a-non-technical-guide-to-interpreting-shap-analyses/


For instance, we see that lower values of % working class have positive SHAP values (the points extending towards the right are increasingly blue) and higher values of % working class have negative SHAP values (the points extending towards the left are increasingly red). This indicates that houses in more working-class areas have lower predicted prices. The reverse is seen for number of rooms - higher room counts lead to higher house price predictions.
